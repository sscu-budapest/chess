```python
from pathlib import Path
import csv
import gzip
import random
import re


import pandas as pd

from tqdm.notebook import tqdm

from src.lichess_decorate import fixt_trepo
from src.lichess import *
```

```python
opath = Path("/share/big/data/path/moves.txt")
```

```python
dza = LichessDza(global_run=True)
```

```python
clean_moves = []
for el in tqdm(dza.project.depot.get_handler_events(GetTournamentGames)):
    for game_str in el.content.decode().strip().split("\n\n\n"):
        _, moves = game_str.split("\n\n")
        parsed_moves = " ".join(re.sub("\d+\.\.\.", "",re.sub("\{.*?\}", "", moves)).split()).replace(".", "")
        if len(parsed_moves) > 100 and len(parsed_moves) < 350:
            clean_moves.append(parsed_moves)
    if len(clean_moves) >= 1_000_000:
        break
```
